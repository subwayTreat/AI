```python
import math

# GOAL TEST FUNCTION
def Goal_Test(n,end):
    if(n==end):
        return 1
    return 0

# Function to get adjacent vertices of a particular vertex
def SUCC(n,graph):
    return graph.get(n)

def euclidean_distance(p1,p2):
    x1,y1 = p1[0],p1[1] 
    x2,y2 = p2[0],p2[1] 
    a = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    return a

def manhatten_distance(p1,p2):
    x1,y1 = p1[0],p1[1] 
    x2,y2 = p2[0],p2[1] 
    a = abs(x2-x1) + abs(y2-y1)
    return a

# Function to calculate euclidean heuristic value between two points
def euclidean_heuristic(p1,p2):
    return euclidean_distance(p1,p2)

```

```python
def best_first_search(graph,location,start,end):
    ol = []       #(node_yet_to_explore,heuristic(node),path_from_start_to_node)
    ol2 = []      #simply to mark visited nodes in open list
    cl = []       #Nodes which are explored
    f={}          #Stores heuristic values
    f[start]=euclidean_heuristic(location[start],location[end])
    p = [start]
    ol.append((start,f[start],p,0))
    ol2.append(start)
    print("At the beginning contents of ol : ",ol,"and contents of closed list : ",cl)
    
    while(ol):
        
        ol.sort(key=lambda a: a[1])      #Sorting based on smallest heuristic value
        curr_node=ol[0][0]
        path=ol[0][2]
        path_cost=ol[0][3]

                
        print("Current node is : ",curr_node)
        ol.pop(0)
        cl.append(curr_node)
        print("Closed List contains : ",cl)
        
        # is goal test true
        if(Goal_Test(curr_node,end)):
            print("Goal Test is True")
            return [path,path_cost]
        
        else:
            print("Goal Test is False")
        
        # finding adjacent vertices of current node 
        s = SUCC(curr_node,graph)
        print("Adjacent vertices of current node : ",s)
        
        for child in s:
            if(child not in ol2 and child not in cl):
                new_path = path + [child]
                h_v = euclidean_heuristic(location[child],location[end])
                h_v = float("{:.2f}".format(h_v))
                f[child]=h_v
                new_path_cost=path_cost+euclidean_distance(location[child],location[curr_node])
                ol.append((child,f[child],new_path,new_path_cost))
                ol2.append(child)
        
        ol.sort(key=lambda a: a[1])
        print("Contents of open list : ",ol)
        print("   ")
        print("------------------------------------------------------")
        print("   ")
    
graph = {'A':['B','D','I'],'B':['A','D'],'C':['D','E','F'],'D':['A','B','C'],'E':['C','F'],
         'F':['C','G','I'],'G':['E','F','J'],'H':['I','J'],'I':['A','F','H'],'J':['H','G']}
location = {'A':[1,1],'B':[2,3],'C':[3,7],'D':[5,5],'E':[5,10],
           'F':[7,8],'G':[10,10],'H':[9,9],'I':[7,1],'J':[11,1]}

ans = best_first_search(graph,location,'A','J')
print("   ")
print("------------------------------------------------------")
print("   ")
print("FINAL path is : ",ans)

# print(euclidean_distance(location['A'],location['I']))
# print(euclidean_distance(location['I'],location['H']))
# print(euclidean_distance(location['H'],location['J']))

```
