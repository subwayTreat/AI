```python
import math

# GOAL TEST FUNCTION
def Goal_Test(n,end):
    if(n==end):
        return 1
    return 0

# Function to get adjacent vertices of a particular vertex
def SUCC(n,graph):
    return graph.get(n)

# Function to calculate euclidean heuristic value between two points
def euclidean_heuristic(p1,p2):
    x1,y1 = p1[0],p1[1] 
    x2,y2 = p2[0],p2[1] 
    a = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    return a

def best_first_search(graph,location,start,end,path):
    ol = []
    cl = []
    path = [start]
    ol.append((start,0,path))
    ol2 = []
    ol2.append(start)
    print("At the beginning contents of ol : ",ol,"and contents of closed list : ",cl)
    
    while(ol):
        
        # finding nearest node and eliminating it from open list
        m = 10**9
        curr_node = 0
        p = []
        for k in ol:
            if(m>k[1]):
                m = min(m,k[1])
                curr_node = k[0]
                p = k[2]
                
        print("Current node is : ",curr_node)
        ol.remove((curr_node,m,p))
        cl.append(curr_node)
        print("Closed List contains : ",cl)
        
        # is goal test true
        if(Goal_Test(curr_node,end)):
            print("Goal Test is True")
            return p
        
        else:
            print("Goal Test is False")
        
        # finding adjacent vertices of current node 
        s = SUCC(curr_node,graph)
        print("Adjacent vertices of current node : ",s)
        
        for child in s:
            if(child not in ol2 and child not in cl):
                new = p + [child]
                h_v = euclidean_heuristic(location[child],location[end])
                h_v = float("{:.2f}".format(h_v))
                ol.append((child,h_v,new))
                ol2.append(child)
                
        print("Contents of open list : ",ol)
        print("   ")
        print("------------------------------------------------------")
        print("   ")
    
graph = {'A':['B','D','I'],'B':['A','D'],'C':['D','E','F'],'D':['A','B','C'],'E':['C','F'],
         'F':['C','G','I'],'G':['E','F','J'],'H':['I','J'],'I':['A','F','H'],'J':['H','G']}
location = {'A':[1,1],'B':[2,3],'C':[3,7],'D':[5,5],'E':[5,10],
           'F':[7,8],'G':[10,10],'H':[9,9],'I':[7,1],'J':[11,1]}
path = []
ans = best_first_search(graph,location,'A','J',path)
print("   ")
print("------------------------------------------------------")
print("   ")
print("FINAL path is : ",ans)
```
